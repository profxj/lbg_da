```python
import numpy as np
from astropy.io import fits
from astropy.table import Table as Table
import matplotlib.pyplot as plt
from linetools.spectra import utils as ltsu
from linetools.spectra.xspectrum1d import XSpectrum1D
import astropy.units as u 
from astropy import constants as const
%matplotlib inline
```

```python
stack2 = fits.open("/home/jsm/PycharmProjects/Summer_Research_2018/SB99/for_sebastien/2.5_data.fits")

stack3 = fits.open("/home/jsm/PycharmProjects/Summer_Research_2018/SB99/for_sebastien/3.0_data.fits")
```

```python
#2<z<2.5
obs_2 = stack2[1].data["FLUX"] # only the data I want 1000-1500A

wave_2 = stack2[1].data["WAVELENGTH"]

sig_2 = stack2[1].data["NOISE"]


#2.5<z<3

obs_3 = stack3[1].data["FLUX"] # only the data I want 1000-1500A

wave_3 = stack3[1].data["WAVELENGTH"]

sig_3 = stack3[1].data["NOISE"]
```

```python
SB99_2 = fits.open("/home/jsm/PycharmProjects/Summer_Research_2018/SB99/for_sebastien/2.5_model.fits")

model_2 = SB99_2[0].data

SB99_3 = fits.open("/home/jsm/PycharmProjects/Summer_Research_2018/SB99/for_sebastien/3.0_model.fits")

model_3 = SB99_3[0].data
```

```python
fig, ax = plt.subplots(figsize=(12,6))

#the spectra
plt.plot(wave_3,obs_3, label="265 LBGs at 2.5 < z < 3.0",color="#4d4d4d")
plt.plot(wave_3,model_3, label = "SB99",color="#b2182b")
plt.plot(wave_3, sig_3,label = "Error", color= "#878787")
ax.axvspan(1070,1170, alpha=0.07, color="c",label="IGM measurements")
plt.xlim(1020,1450)


#interstellar absorption features
plt.text(1260.42,.25,"$SiII$",fontsize=12,color="#d95f0e")

plt.text(1303.1685,.25,"$SiII$",fontsize=12,color="#d95f0e")

plt.text(1334.5323,.25,"$CII$",fontsize=12,color="#d95f0e")

plt.text(1390,.25,"$SiIV$ doublet",fontsize=12,color="#d95f0e")

plt.text(1176,.25,"$CIII$",fontsize=12,color="#d95f0e")

#stellar absorption features
plt.text(1202,.25,"$SiIII$",fontsize=12,color="#d95f0e")

plt.text(1083,.25,"$NII$",fontsize=12,color="#d95f0e")

plt.text(1122,.25,"$FeIII$",fontsize=12,color="#d95f0e")

plt.text(1141,.25,"$NI$",fontsize=12,color="#d95f0e")

#emission features
plt.text(1219,2.0,"$Ly$",fontsize=12,color="#d95f0e")

#the axis and title
plt.title("LBG Composite $z_{median}$ = 2.657",fontsize=20)
plt.ylabel("Relative Flux",fontsize=18)
plt.xlabel("Wavelength ",fontsize=18)
plt.axhline(0,ls="--",color="black")



#the limits
plt.ylim(-.25,3.0)
plt.xlim(1050,1450)

plt.legend()
plt.show()

fig.savefig("/home/jsm/my_lbg_da/lbg_da/summer_2018/figures/png/LBG_3.0.png",bbox_inches='tight')
```
