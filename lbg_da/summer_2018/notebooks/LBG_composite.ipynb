```python
import numpy as np
from astropy.io import fits
from astropy.table import Table as Table
import matplotlib.pyplot as plt
from linetools.spectra import utils as ltsu
from linetools.spectra.xspectrum1d import XSpectrum1D
import astropy.units as u
from astropy import constants as const
%matplotlib inline
```

```python
#2<z<2.5
stack_2 = XSpectrum1D.from_file("/home/jsm/PycharmProjects/Summer_Research_2018/lbg_da/lbg_da/summer_2018/notebooks/composite_2.0.fits")
#2.5<z<3
stack_3 = XSpectrum1D.from_file("/home/jsm/PycharmProjects/Summer_Research_2018/lbg_da/lbg_da/summer_2018/notebooks/composite_3.0.fits")

flux_3 = stack_3.flux
wave_3 = stack_3.wavelength

flux_2 = stack_2.flux
wave_2 = stack_2.wavelength
```

```python
fig, ax = plt.subplots(figsize=(16,6))

#the spectra
plt.plot(wave_2,flux_2,label="177 LBGs at 2.0 < z < 2.5")#,color="r")
plt.plot(wave_3,flux_3,label="136 LBGs at 2.5 < z < 3.0")#,color="orange")
ax.axvspan(1070,1170, alpha=0.07, color="green",label="IGM measurements")


#stellar absorption features
plt.axvline(1108,ls=":",label="Stellar absorption features",color="cyan")
plt.text(1108,-1,"CIV",fontsize=12)

plt.axvline(1206,ls=":",color="cyan")
plt.text(1206,-1,"SiIII",fontsize=12)


#interstellar absorption features
plt.axvline(1260.42,ls=":",label="Interstellar absorption features",color="green")
plt.text(1260.42,-1,"SiII",fontsize=12)

plt.axvline(1303.1685,ls=":",color="green")
plt.text(1303.1685,-1,"OI + SiII",fontsize=12)

plt.axvline(1334.5323,ls=":",color="green")
plt.text(1334.5323,-1,"CII",fontsize=12)

plt.axvline(1393,ls=":",color="green")
plt.text(1393,-1,"SiIV",fontsize=12)

plt.axvline(1402,ls=":",color="green")
plt.text(1402,-.5,"SiIV",fontsize=12)

plt.axvline(1175,ls=":",color="green")
plt.text(1175,-1,"CIII",fontsize=12)


#emission features
plt.axvline(1216.67,ls=":",color="red",label="Emission features")
plt.text(1216.67,-.5,"Ly",fontsize=12)

plt.axvline(1309.276,ls=":",color="red")
plt.text(1309.276,-.5,"SiII",fontsize=12)


#the axis and title
plt.title("LBG Composite Spectra",fontsize=20)
plt.ylabel("Flux",fontsize=18)
plt.xlabel("Wavelength ",fontsize=18)
plt.axhline(0,ls="--",color="black")

#the limits
plt.ylim(-1.5,9)
plt.xlim(1050,1400)

plt.legend()
plt.savefig("LBG_composite.pdf")
plt.show()
```
