```python
import numpy as np
from astropy.io import fits
from astropy.table import Table as Table
import matplotlib.pyplot as plt
from linetools.spectra import utils as ltsu
from linetools.spectra.xspectrum1d import XSpectrum1D
import astropy.units as u
from astropy import constants as const
%matplotlib inline
```

```python
#2<z<2.5
stack_2 = XSpectrum1D.from_file("/home/jsm/my_lbg_da/lbg_da/summer_2018/notebooks/composite_2.5.fits")
#2.5<z<3
stack_3 = XSpectrum1D.from_file("/home/jsm/my_lbg_da/lbg_da/summer_2018/notebooks/composite_3.0.fits")

flux_2 = stack_2.flux
wave_2 = stack_2.wavelength

flux_3 = stack_3.flux
wave_3 = stack_3.wavelength
```

```python
fig, ax = plt.subplots(figsize=(16,6))

#the spectra
plt.plot(wave_2,flux_2,label="281 LBGs at 2.0 < z < 2.5",color="red")
plt.plot(wave_3,flux_3,label="226 LBGs at 2.5 < z < 3.0",color="cornflowerblue")
ax.axvspan(1070,1170, alpha=0.07, color="gold",label="IGM measurements")


#stellar absorption features
plt.axvline(1208,ls=":",color="pink",label="Stellar absorption features")
plt.text(1208,-1,"SiIII",fontsize=12)


#interstellar absorption features
plt.axvline(1260.42,ls=":",label="Interstellar absorption features",color="c")
plt.text(1260.42,-1,"SiII",fontsize=12)

plt.axvline(1303.1685,ls=":",color="c")
plt.text(1303.1685,-1,"OI + SiII",fontsize=12)

plt.axvline(1334.5323,ls=":",color="c")
plt.text(1334.5323,-1,"CII",fontsize=12)

plt.axvline(1393,ls=":",color="c")
plt.text(1393,-1,"SiIV",fontsize=12)

plt.axvline(1402,ls=":",color="c")
plt.text(1402,-.5,"SiIV",fontsize=12)

plt.axvline(1176,ls=":",color="c")
plt.text(1176,-1,"CIII",fontsize=12)


#emission features
plt.axvline(1216,ls=":",color="violet",label="Emission features")
plt.text(1216,-.5,"Lyα",fontsize=12)

plt.axvline(1309.276,ls=":",color="violet")
plt.text(1309.276,-.5,"SiII",fontsize=12)


#the axis and title
plt.title("LBG Composite Spectra",fontsize=20)
plt.ylabel("Flux",fontsize=18)
plt.xlabel("Wavelength Å",fontsize=18)
plt.axhline(0,ls="--",color="black")

#the limits
plt.ylim(-1.5,9)
plt.xlim(1050,1430)

plt.legend()
plt.savefig("LBG_composite.pdf")
plt.show()
```
