```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
```

```python
tau_data = np.load("../data/tau_eff/tau_data.npy")
```

```python
lowz = 2.442 

hiz =  2.606

red = tau_data[0,:]
```

```python
#z_i = (wave/ 1216) * (1 + z_med) - 1

wave = np.linspace(1070,1170,len(red))
```

```python
def zmed(wave, z_i):
    
    return (z_i + 1)/(wave/1216) - 1
```

```python
hmm = np.array(zmed(wave, red))
```

```python
plt.hist(hmm)
plt.vlines(lowz,0,30)
plt.vlines(hiz,0,30)
```
