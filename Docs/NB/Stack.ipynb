# A function for stacking a list of Xspectrum1D 

```python
def stack_xpsec( speclist, zvalues ):
    
    # Collate the list into one object
    collate = ltsu.collate(speclist)
    
    # Create an array from the list of z values
    zarray = np.array((zvalues))       #need to be in order according to the speclist
    
    # Shift the object to the rest frame
    rspecs = ltsu.rebin_to_rest(collate, zarray, 200*u.km/u.s) 
    
    for i in range((rspecs.nspec)):
        rspecs[i].plot()
            
    # Stack the spectra
    n_stack = ltsu.smash_spectra(rspecs)
    
    n_stack.plot()
    
    return n_stack
        
```
